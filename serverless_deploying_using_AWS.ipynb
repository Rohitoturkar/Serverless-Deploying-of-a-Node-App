### SSH
to get ssh access of any ecw instace using ssh cli the steps are
1. the RSA security key which you have created at time of creation of ec2 instance and want to gain access via ssh should be use here , you will navigate to the folder where you have stored it using the linux cli and then copy the ssh command present on ec2 ssh client and run it on your linux cli


## How to do it 

1. create a ec2 instance for building a dockerfile and cloning your github code  

2. now we haave to push the dockerfile on ecr(elastic container repo) therefore we will first create a repo in ECR with our desired name , now we have created repo ,to push the image in it we will view the push commands used to push the image in it 



3. do whatever are necessary inside the ec2 as asked in the push commands
4. after installing docker , we have to give the user permission to the user by using __$sudo usermod -aG docker $user__ this gives the user the permission to use docker 
5. install other necessities mentioned in the push commands 
6. now retreive a token and authenticate the docker client , for this we will have to create a user using IAM user , this user will have access key for command line ,and also provide him with ECR permission and give full access 
7. finally run the command to build the docker image 
8. run the final command to push it into ecr

9. now to create a fully managed container we will use __ECS__ elastic container service




we create the ecs for enabling auto scalling of our docker container according to the traffic , for that we need to create a task definition , it contains all the necessary info men tioned in the image 

10. in ECS select __fargate__ which will help to create dynamic container , and then click on create cluster
11. now add tasks in that cluster , creating new task by creating task definition
12. select how much cpu and memory you want it to have at minimum to maintained at all time for smooth workflow 
13. select task execution role as __ECstaskexecution_ROle__
14. provide the URL of the image created in ECR and paste it into task definition so that it can run it
15. map the port you want it to run on it
16. select the max number of cpu you want it to have 
17. now run the task definition in the cluster in ECS, select the cluster you have to run it on 
18. finally the task definition is created 
the flow is create a cluster--> create a task definition--> assign the task definition to the cluster

19. now all is set , go to the cluster , click on the task you will get a public ip from where the application can be accessed
